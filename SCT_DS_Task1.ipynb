# Extraction of top 10 countries with respect to total population

```python
# filter data for total population
total_pop_data = df[ df ['Series Code'] == 'SP.POP.TOTL']

#sort data based on the total population for 2022
total_pop_sorted = total_pop_data.sort_values(by = '2022', ascending=False)

#get the top 10 countries with the highest total population
total_top_ten_country = total_pop_sorted.head(10)

print("Top Ten Countries od total population \n")
print(total_top_ten_country[['Country Code']])
```

# Bar Plot

# Top 10 countries of total population in year 2022 and 2016

```python
# Create a bar plot

plt.figure(figsize = (15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=total_top_ten_country,palette='coolwarm')
plt.title('Top 10 Countries of total Population(2022)',fontsize = 10)
plt.xlabel("Total Population",fontsize = 10)
plt.show()

plt.figure(figsize = (15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=total_top_ten_country,palette='coolwarm')
plt.title('Top 10 Countries of total Population(2016)',fontsize = 10)
plt.xlabel("Total Population",fontsize = 10)
plt.show()
```

# Extraction of top 10 countries with highest male population

```python
# filter data for male population

male_population_data = df[df['Series Code']=='SP.POP.TOTL.MA.IN']

# sort data based on male population based on 2022
male_pop_sorted = male_population_data.sort_values(by = '2022', ascending = False)

# top 10 countries with the highest male population in 2022

male_top_ten_country = male_pop_sorted.head(10)

print('Top 10 countres of male Population')
print(male_top_ten_country[['Country Code']])
```

```python
# filter data for female population

female_population_data = df[df['Series Code']=='SP.POP.TOTL.FE.IN']

# sort data based on male population based on 2022
female_pop_sorted = female_population_data.sort_values(by = '2022', ascending = False)

# top 10 countries with the highest male population in 2022

female_top_ten_country = female_pop_sorted.head(10)

print('Top 10 countres of female Population')
print(female_top_ten_country[['Country Code']])
```

# Bar PLot

# Top 10 countries with Highest male and female population in 2022

```python
plt.figure(figsize = (15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=male_top_ten_country,palette='viridis')
plt.title('Top 10 countries of male population(2022)',fontsize = 10)
plt.xlabel("Male Population",fontsize = 10)
plt.show()

plt.figure(figsize = (15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=female_top_ten_country,palette='viridis')
plt.title('Top 10 countries of female population(2022)',fontsize = 10)
plt.xlabel("Female Population",fontsize = 10)
plt.show()
```
